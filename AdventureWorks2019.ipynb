## 5]Questions & Answers

```python
# What is the total sales for each product?
df.groupby('ProductName').sum()['TotalDue']
```

```python
# What is the avergae cost by shipping?
df.groupby('ShipMethodName')['Freight'].mean()
```

```python
#What is the total sales for each territory?
df.groupby('TerritoryName').sum()['TotalDue']
```

```python
#What is the total number of products in each category?
df.groupby('CategoryName').count()['ProductName']
```

```python
#What is the average purchase for each category?
df.groupby('CategoryName')['TotalDue'].mean()
```

```python
#What is the total number of products shipped by each method?
df.groupby('ShipMethodName').count()['ProductID']
```
